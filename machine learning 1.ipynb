# What is machine learning?
Machine learning (ML) is a branch of artificial intelligence (AI) and computer 
science 
that focuses on the using data and algorithms to enable AI to imitate the way that humans
 learn, gradually improving its accuracy.


***Structured data*** is data that fits neatly into data tables and includes discrete data 
types such as numbers, short text, and dates. ***Unstructured data*** doesn't fit neatly into
a data table because its size or nature: for example, audio and video files and large text
documents.

***Main types of ML***
1-Supervised
2-unsupervised
3-Transfer
4-Reinforcement



# 1-Supervised learning
Supervised learning is a type of machine learning where an algorithm is trained on labeled data. Labeled data means that each input comes with a correct output or "label" that the model is supposed to predict.

For example, imagine you have a bunch of pictures of animals. Each picture is labeled with the animal's name, like "cat" or "dog." The algorithm learns from these examples so that when you show it a new picture, it can guess whether the picture is of a cat, dog, or another animal.

In short, it's like teaching a computer by showing it examples and their correct answers, so it learns to predict the right answer for new data.
***Classification*** and ***regression*** are two types of problems you solve using supervised learning.

Classification and regression are two types of problems you solve using supervised learning. Here's a simple explanation:

### **Classification**:
- **Goal**: To predict **categories** or **classes**.
- **Example**: Imagine you're given pictures of animals, and your goal is to classify each picture as either a "cat" or a "dog."
- **Output**: A **discrete** value (like "cat" or "dog"). It's about putting things into groups or categories.

**Another example**: Sorting emails as "spam" or "not spam" is a classification problem.

### **Regression**:
- **Goal**: To predict a **numerical value**.
- **Example**: Suppose you want to predict the price of a house based on features like its size and location.
- **Output**: A **continuous** value (like a price, temperature, or age). It's about predicting numbers.

**Another example**: Predicting someone's weight based on their height and age is a regression problem.

In summary:
- **Classification**: Predicting **categories** (e.g., cat or dog).
- **Regression**: Predicting **numbers** (e.g., house price).

# 2-Unsupervised learning

Unsupervised learning is when a computer learns from **data that doesn't have answers** or labels. The goal is to find patterns in the data on its own.

### Two main things it does:
1. **Clustering**: It groups similar things together. For example, it can group people who like similar movies, even though no one told it which movies belong together.
   
2. **Dimensionality Reduction**: It helps simplify data by removing unnecessary details while keeping important information.

Basically, unsupervised learning is like exploring a new place without a map and trying to figure out the best way around.

# 3- Transfer learning

Transfer learning is when a model that has been trained for one task is used as a starting point for a new, but related task. Instead of training a model from scratch, you take an already-trained model and adapt it to the new task, which saves time and resources.

### Example:
Imagine you have a model that can recognize objects in pictures (like cats, dogs, and cars). You can take this model and use it to help recognize something similar, like different types of flowers. The model already knows a lot about shapes and features, so you just need to fine-tune it a little for the new task.




# 4- Reinforcement learning

Reinforcement learning is like teaching a computer by **giving it rewards** for doing the right thing and **penalties** for doing the wrong thing. Over time, it learns to make better choices to get more rewards.

### Example:
Imagine you’re playing a game. When you win, you get points (reward). If you lose, you get no points (penalty). You keep playing and learn what moves help you win. The computer does the same—it tries things, learns what works, and gets better at the task over time.

It's like trial and error until it figures out the best way to succeed.

### Matching your problems

***Supervised***:I know inputs and outputs

***Unsupervised***:I'm not sure of the outputs but I have inputs

***Transfer***:I think my problem may be similar to something else


# Modelling

***Training***-> train your model on this

***Validation*** -> Tune your model on this

***Test*** -> test and compare on this

***Example***:

Imagine you're studying for a test. You do practice questions (training), but then you check your knowledge by doing some extra questions that you haven’t seen before (validation). This shows you how well you're really learning the material, not just memorizing the answers.



***Overfitting***: This happens when a model learns too much from the training data, including noise and details that aren’t important. It performs well on the training data but poorly on new, unseen data.

***Underfitting***: This happens when a model doesn’t learn enough from the training data. It’s too simple and misses key patterns, so it doesn’t perform well on both training data and new data.

The goal is to find the right balance, where the model learns enough to work well on new data without memorizing unnecessary details.
